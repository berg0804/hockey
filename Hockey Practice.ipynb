```python
import os
import pandas as pd
import datetime as dt
import numpy as np
```

```python
playoffs2022_load = os.path.join("Files", "Playoffs_2022.csv")
```

```python
playoffs2022_df = pd.read_csv(playoffs2022_load)
```

```python
playoffs2022_df['xgdiff'] = playoffs2022_df["xGA"] - playoffs2022_df["GA"]
```

```python
playoffs2022_df['gdiff'] = playoffs2022_df["GF"] - playoffs2022_df["GA"]
```

```python
playoffs2022_df.loc[playoffs2022_df['Result'] > 0, 'Team Win/Loss'] = '1' 
playoffs2022_df.loc[playoffs2022_df['Result'] < 1, 'Team Win/Loss'] = '0' 
```

```python
playoffs2022_df
```

```python
playoffs2022_df['Team Win/Loss'] = playoffs2022_df['Team Win/Loss'].astype(int)
```

```python
conditions = [
    (playoffs2022_df['Team Win/Loss'] < 1) & (playoffs2022_df['xgdiff'] < playoffs2022_df['gdiff']),
    (playoffs2022_df['Team Win/Loss'] > 0) & (playoffs2022_df['xgdiff'] > playoffs2022_df['gdiff']),
    (playoffs2022_df['Team Win/Loss'] > 0) & (playoffs2022_df['xgdiff'] < 0),
    (playoffs2022_df['Team Win/Loss'] < 1) & (playoffs2022_df['xgdiff'] > 0)
]

values = ['Goalie Loss'], ['Goalie Win'], ['Bad in Win'], ['Strong in Loss']

playoffs2022_df['Goalie Win/Loss'] = np.select(conditions, values)
                
playoffs2022_df
```

```python
summary_df = playoffs2022_df.drop(['Date','Position', 'TOI','GP', 'GF', 'GA', 'GF%', 'xGF', 'xGA', 'xGF%', 'Team', 'Result', 'xgdiff', 'gdiff', 'Team Win/Loss'], axis='columns')
summary_df.head()
```

```python
summarytable_df = summary_df.pivot_table(index='Player', columns='Goalie Win/Loss', aggfunc=len)
summarytable_df.head()
```
